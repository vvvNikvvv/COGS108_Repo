# Research Question

What are the factors associated with performance ratings of clinicians under the Merit-based Incentive Payment System (MIPS)? Specifically, what are the common characteristics of clinicians who perform exceptionally well (MIPS score ≥ 85) compared to those who receive neutral (60 ≤ MIPS score < 85) or negative (MIPS score < 60) scores?

We will examine factors such as:

Demographic Factors: Gender, medical school name, and year of graduation (experience).
Practice Information: Primary specialty, telehealth offering, location (state), type of facility, type of procedures performed, and frequency of procedures performed.
Performance Metrics: Overall MIPS score, Quality category score, Promoting Interoperability (PI) category score, and Improvement Activities (IA) category score.

## Background and Prior Work

The Merit-based Incentive Payment System (MIPS) was established as part of the Medicare Access and CHIP Reauthorization Act (MACRA) of 2015 to transition Medicare from a fee-for-service model to a value-based care system3. This shift was necessitated by the inefficiencies and high costs associated with the traditional fee-for-service approach, which often prioritized volume over quality of care8. MACRA introduced the Quality Payment Program (QPP), which includes two tracks: MIPS and Advanced Alternative Payment Models (APMs), aiming to incentivize high-quality, cost-efficient care through financial bonuses and penalties1.

MIPS consolidated several legacy programs (PQRS, VBM, and the EHR Incentive Program) into a single framework2. The introduction of MIPS was a pivotal step toward enhancing transparency and accountability in healthcare. By publicly reporting performance scores, MIPS provides patients with valuable information about clinician quality, helping them make informed healthcare decisions5. For clinicians, MIPS incentivizes continuous improvement in care delivery and the adoption of best practices4.

There have been numerous research studies that examined the impact of implementing MIPS into the healthcare system. Berkowitz et al. (2022) evaluated the economic impact of MIPS on ophthalmologists, revealing that 92.9% of participants met criteria for positive payment adjustments, with a significant proportion achieving exceptional performance scores. However, ophthalmologists filing as individuals were less likely to achieve these scores compared to those filing as groups. This suggests that a clinician’s affiliation with a particular facility may offer structural advantages that facilitate higher performance, suggesting a potential flaw in determining the MIPS score12.

Cheng et al. (2020) also supports the notion that MIPS has positively influenced the performance of clinicians. This study has noted that nephrologists participating in the system saw improved care quality and cost savings6. However, the transition to MIPS posed unique challenges as well, particularly for specialties with fewer applicable performance metrics9 11. Jones et al. (2018) found that oncologists had to adapt significantly to meet MIPS requirements due to the types of procedures they have to perform compared to clinicians in other fields, which emphasized the discrepancy in performance measurement by each specialty7.

Early data from the implementation of MIPS indicated high participation rates among eligible clinicians, with 91% engaging in the QPP in 20178. However, MIPS has been criticized for its complexity in score calculations and the significant reporting burden it places on clinicians, suggesting that the program might not effectively drive quality improvements11. Similar to Jones et al. (2018), Gettel et al. (2021) found that emergency clinicians faced particular challenges in meeting MIPS requirements as they often had to perform various spontaneous tasks without prior engagement with the patients. This study found the scoring system did not align well with the realities of emergency care, again suggesting flaws in the system that pertain to a particular specialty10.

While these studies provide valuable insights into the economic impacts and operational challenges associated with MIPS, there remains a gap in comprehensively understanding the factors that contribute to varying MIPS performance scores across different clinicians. This study aims to fill that gap by examining a range of variables, including gender, credential, medical school, year of graduation, telehealth offering, and location along with known factors related to MIPS score such as specialty of the clinician, facility they work in, and types and frequencies of procedures performed, to identify determinants of clinician performance under MIPS and further point to improvement in the scoring matrix. These insights can better patient outcomes, inform policy and program development, enhance clinician education, support telehealth integration, and promote equity in healthcare, thereby supporting the overarching goals of value-based care initiatives.

The Exploratory Data Analysis of the dataset provides a broad view of the factors potentially influencing MIPS scores among clinicians:

1. **Specialization Distribution**: The first notable observation is the distribution across specialties, with "Orthopedic Surgery," "General Surgery," and "Cardiology" among the most common. This specialization data, coupled with MIPS scores, could indicate that certain specialties either due to the complexity or volume of procedures, achieve differing levels of MIPS performance. This aligns with the hypothesis that specialty affects MIPS scores.

2. **Procedural Frequency**: The types of procedures frequently conducted (e.g., colonoscopy, hip replacement) suggest a correlation between high-frequency procedures and the specialties. This may impact MIPS scores as specialties that perform more complex or frequent interventions might be better equipped or have more opportunities to score higher on performance metrics.

3. **Geographic Distribution**: The data on clinicians' distribution by state and the locations of their medical schools suggests geographical diversity. However, further analysis would be required to determine if state-specific healthcare policies or the concentration of medical services in certain states influence MIPS scores.

4. **Facility Type**: Over 90% of clinicians work in hospitals, which are typically equipped with better resources and infrastructure than other facility types like home health agencies or nursing homes. This could suggest that hospital-based clinicians have a different performance profile on MIPS, likely due to the complex and varied cases handled in such settings.

5. **Demographics (Gender and Graduation Year)**: The histograms show a relatively even distribution of genders and a recent trend towards younger clinicians. While the analysis does not directly measure the impact of these factors on MIPS scores, the data supports a demographic shift in the workforce, which could influence new policies and training that reflect more recent medical education trends and gender diversity.

6. **Primary Specialty vs. Graduation Year**: Scatter plots examining the relationship between primary specialty and graduation year do not show any clear trends or clusters, suggesting that while these variables are important, their influence on MIPS scores may be less direct or require more nuanced statistical methods to uncover any patterns.

The EDA supports the hypothesis to an extent by highlighting that certain professional factors such as specialty and the type of facility where clinicians work could influence MIPS scores. However, the impact of demographic factors like gender and graduation year is less clear from the basic visual analysis provided and would benefit from more detailed statistical testing such as regression analysis or machine learning models to quantify these effects.



# Ethics & Privacy

The research question focuses on identifying factors associated with performance ratings in clinicians by MIPS, raising ethical concerns such as ensuring the analysis does not unfairly disadvantage groups based on characteristics like gender, age, or location. Data collection will use publicly available sources from CMS and AMA datasets. Ethical considerations during data collection include protecting individual clinician identities and ensuring that consent protocols were followed during the original data collection.

Reporting bias might also occur if clinicians underreport or overreport certain performance metrics, and systematic bias may reflect historical biases in healthcare, such as gender biases in specialty choice.

During the analysis, ensuring fair representation and avoiding the perpetuation of existing biases are essential. The analysis should consider the equitable impact on different groups, especially regarding how MIPS scores and subsequent policy recommendations affect them. Post-analysis concerns include protecting individual privacy when sharing results and ensuring policy recommendations do not unfairly impact any group of clinicians.

To handle these biases and ethical concerns, exploratory data analysis will be conducted to identify any apparent biases, such as unequal representation of different demographic groups. Although the study is exploratory and adjustments for bias are not planned as causal claims will not be made from the analysis, any identified biases will be noted. It is important to note that the dataset may be an unrepresentative sample that cannot be generalized to the entire population. This limitation will be clearly stated in the findings. The analysis will consider the context of different groups, such as varying access to resources or differing patient populations, to ensure equitable analysis. Transparency about the methodology and assumptions used in the analysis will be maintained, with clear documentation and rationales provided for all decisions made during the data science process.
